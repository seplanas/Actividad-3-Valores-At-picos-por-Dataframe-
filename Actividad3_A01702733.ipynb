```python
import pandas as pd
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name = '2023', header = 3 )
df
```

```python
df_nulos = df.isnull().sum()
df_nulos
```

```python
#Insertar valores nulos en columnas categóricas
df_folio = df["Folio"] =df["Folio"].fillna(method="ffill")
df_gasto = df["MP"] =df["MP"].fillna(method="ffill")
df_poliza = df["Poliza"] =df["Poliza"].fillna(method="ffill")
df_OTROS = df["Otros"] =df["Otros"].fillna(method="ffill")
df_tipo = df["TIPO GASTO"] =df["TIPO GASTO"].fillna(method="ffill")
```

```python
#sustituir valores numéricos
df_TC = df["FP"]=df["FP"].fillna(round(df["TC"].mean()))
df_importe = df["Importe"]=df["Importe"].fillna(round(df["Importe"].mean()))
```

```python
df_nulos = df.isnull().sum()
df_nulos
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (15,8))
df.plot(kind="box", vert= False)
plt.suptitle("Valores atípicos")
plt.show()
```

```python
limite_superior = df.mean() + 3*df.std()
limite_inferior = df.mean() - 3*df.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

```python
#cambiar valores atípicos a nulos
df2 = df[(df <= limite_superior) & (df >=limite_inferior)]
df2
```

```python
df3 = df2.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO","Descripción","Tipo", "Status", "Poliza","Otros"], axis=1)
df3
```

```python
#corroboramos los valores nulos
df3.isnull().sum()
```

```python
#sustituyo con valores de media
df_clean = df3.copy()
df_clean = df_clean.fillna(round(df3.mean(),1))
df_clean
```

```python
df_clean.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df_clean)
plt.suptitle("Outliers desviación estándar")
plt.show()
```

```python
per25 = df_clean.quantile(0.25) #Q1
per75 = df_clean.quantile(0.75) #Q3

iqr = per75 - per75

lim_sup_iqr = per75+ 1.5*iqr
lim_inf_iqr = per25- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print("\nLímite inferior permitido" , lim_inf_iqr)
```

```python
#datos limpios de df
df3_iqr = df_clean[(df_clean <=lim_sup_iqr)&(df_clean >=lim_inf_iqr)]
df3_iqr
```

```python
df4_iqr = df3_iqr.copy()
df4_iqr = df4_iqr.fillna(round(df3_iqr.mean(),1))
df4_iqr
```

```python
#unimos columnas string eliminadas
columnas = ["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO","Descripción","Tipo", "Status", "Poliza","Otros"]
indice= df[columnas]
datos_limpios = pd.concat([indice, df4_iqr],axis=1)
datos_limpios
```
